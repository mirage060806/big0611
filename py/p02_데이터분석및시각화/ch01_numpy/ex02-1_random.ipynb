```python
import random

# 1부터 45까지의 숫자 중 하나를 무작위로 추출 (45 포함!)
lotto_single = random.randint(1, 45)
print(lotto_single)
```

```python
# 0부터 9까지(10 직전)의 짝수 중 하나를 무작위로 추출
even_num = random.randrange(0, 10, 2)
print(even_num)
```

```python
# 0.0이상~1.0미만
prob = random.random()
print(prob)  # 예: 0.3547...

# 응용: 30% 확률 구현하기
if prob < 0.30:
    print("강화 성공!")
else:
    print("강화 실패...")
```

```python
# uniform(a, b): a 이상 b 이하의 실수 1개 추출
```

```python
print(random.uniform(1.5, 5.5))  # 예: 3.742...
```

```python
# 리스트(시퀀스) 데이터 다루기
```

```python
## choice(sequence): 목록에서 무작위로 1개 고르기
```

```python
menu = ["짜장면", "짬뽕", "탕수육", "피자"]
today_lunch = random.choice(menu)
print(today_lunch)  # 메뉴 중 하나가 무작위 출력됨
```

```python
# sample(sequence, k): 중복 없이 여러 개 고르기
```

```python
numbers = list(range(1, 46))
# 1~45 중에서 중복 없이 6개 번호 추첨
lotto_numbers = random.sample(numbers, 6)
print(lotto_numbers)  # 보기 좋게 정렬해서 출력
```

```python
# shuffle(sequence): 목록을 무작위로 뒤섞기
```

```python
cards = ["A", "K", "Q", "J"]
random.shuffle(cards)
print(cards)  # 원본 순서가 뒤섞인 결과 출력 (예: ['Q', 'A', 'J', 'K'])
```
