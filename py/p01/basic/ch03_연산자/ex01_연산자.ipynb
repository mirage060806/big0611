### 연산자
#### 1. 산술 연산자

```python
a = 10
b = 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a ** b)
print(a % b)
```

2. 비교 연산자
- \>
- <
- \>=
- <=
- ==
- !=

```python
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
print(a == b)
print(a != b)
```

#### 3. 논리 연산자
- and(논리곱)
- or(논리합)
- not(논리부정)

```python
print((10 > 5) and (3 < 1) and (5 > 1))
print((10 > 5) or (3 < 1) and (4 < 1))
print(not(10 > 5))
```

#### 4. 대입 연산자

```python
# 5를 x에 대입
# 오른쪽 값을 왼쪽에 대입
# 변수 x, a, b
x = 5
# 5 = x -> SyntaxError
# b의 값을 a에 대입
a = b
```

#### 5. 복합 대입 연산자
- +=
- -=
- *=
- /=
- %=
- //=
- **=

```python
x = 5
x += 3
print(x)
x -= 3
print(x)
x *= 3
print(x)
x /= 3
print(x)
x %= 3
print(x)
x //= 3
print(x)
x **= 3
print(x)
```

#### 6. 멤버십 연산자
- in: 값이 안에 있으면 True
- not in: 값이 안에 없으면 True

```python
fruits = ["apple", "banana", "cherry"] # 리스트
print("banana" in fruits)
print("orange" in fruits)
```
